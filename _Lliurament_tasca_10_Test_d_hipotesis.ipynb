```python
# L'atribut a analitzar és 'turns', és a dir, nombre de movimens
moviments         = list(df_partides_escacs.turns)

# Paràmetres del test
mu                = 60
alfa              = 0.05

# Cácul del valor del test tstudent observació 'tobservacio'
# Cálcul de diferents estadístics de 'turns' i mida de mostra 'n'
moviments_mitjana = stat.mean(moviments)
moviments_desv    = stat.stdev(moviments)
n                 = len(moviments)

t_observacio      = (moviments_mitjana-mu)/(moviments_desv/np.sqrt(n))

# Interval d'acceptació inferior i superior
interval_inferior = t.ppf(alfa/2,n-1)
interval_superior = t.ppf(1-alfa/2,n-1)

# Cálcul del p_valor
p_valor           = t.sf(abs(t_observacio),n-1)*2

print("Els valors dels diferents paràmetres y estadístics calculats son:\n\nt_observacio --------> {}\ninterval inferior  --> {}\ninterval superior ---> {}\np_valor -------------> {}\n".format(round(t_observacio,4),round(interval_inferior,4),round(interval_superior,4),round(p_valor,4)))
```

###### El valor observat cau fora del interval d'aceptació i el p_valor es menor que el nivell de significancia alfa=0.05. Per tant, podem dir amb un 95% de confiança que la mitjana de moviments és diferent de 60

```python
# Paràmetres del test
moviments_llindar = 45
alfa              = 0.05

# L'atribut a analitzar és 'turns', és a dir, nombre de movimens i diferenciem blanques de negre
moviments_blanques= df_partides_escacs[(df_partides_escacs['winner']=='white')&(df_partides_escacs['turns']>moviments_llindar)].turns
moviments_negres  = df_partides_escacs[(df_partides_escacs['winner']=='black')&(df_partides_escacs['turns']>moviments_llindar)].turns

# Cácul del valor del test z observació 'z_observacio'
n_blanques        = moviments_blanques.shape[0]
n_negres          = moviments_negres.shape[0]
pctg_blanques     = sum(moviments_blanques)/n_blanques
pctg_negres       = sum(moviments_negres)/n_negres

p_comu            = (n_blanques*pctg_blanques+pctg_negres*pctg_negres)/(n_blanques+n_negres)
z_observacio      = (pctg_blanques-pctg_negres)/np.sqrt(p_comu*(100-p_comu)*(1/n_blanques+1/n_negres))

# Interval d'acceptació inferior i superior
interval_inferior = norm.ppf(alfa/2)
interval_superior = norm.ppf(1-alfa/2)

# Cálcul del p_valor
p_valor           = norm.sf(abs(z_observacio))*2

print("Els valors dels diferents paràmetres y estadístics calculats son:\n\nz_observacio --------> {}\ninterval inferior  --> {}\ninterval superior ---> {}\np_valor -------------> {}\n".format(round(z_observacio,4),round(interval_inferior,4),round(interval_superior,4),round(p_valor,4)))
```
