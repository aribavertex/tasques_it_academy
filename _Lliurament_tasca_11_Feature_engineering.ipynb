```python
import pandas as pd
pd.options.display.max_columns=None
from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
# Carrreguen i convertim a dataframe la taula continguda dins el fitxer/conjunt de dades anomenat games.csv
df_partides_escacs=pd.read_csv('H:\_vertex\_formacio\_python - cibernarium\_07\_07_11\games.csv',encoding="ISO-8859-15")
```

```python
# Visualització dels 5 primers registres del conjunt de dades
display(df_partides_escacs.head(5))
```

```python
# Estructura del conjunt de dades: nom, valors no nuls i tipus de columna
df_partides_escacs.info()
```

```python
# Estadístics del conjunt de dades
df_partides_escacs.describe(include="all")
```

```python
# Generar els dummis de les variables categóriques amb pocos categories
df_partides_escacs_victory_status_dummies=pd.get_dummies(df_partides_escacs["victory_status"])
df_partides_escacs_winner_dummies=pd.get_dummies(df_partides_escacs["winner"])

# Afegir columnes noves al nostre conjunt de dades
df_partides_escacs=pd.concat([df_partides_escacs,df_partides_escacs_victory_status_dummies,df_partides_escacs_winner_dummies],axis=1)

# Visualització dels 5 primers registres del conjunt de dades
display(df_partides_escacs.head(5))
```

```python
# Estandarització d'algunes columnes numériques: 'turns','white_rating' i 'black_rating'
estandaritzar=StandardScaler()
df_partides_escacs_columnes_standaritzades=['turns_escalat','white_rating_escalat','black_rating_escalat']
df_partides_escacs[df_partides_escacs_columnes_standaritzades]=estandaritzar.fit_transform(df_partides_escacs[['turns','white_rating','black_rating']].values)

# Visualització dels 5 primers registres del conjunt de dades
display(df_partides_escacs.head(5))
```

```python
# Creació del dataframe per a fer l'anàlis de components principals (ACP)
df_partides_escacs_columnes_acp=pd.concat([df_partides_escacs[df_partides_escacs_columnes_standaritzades]],axis=1)

# Visualització dels 10 primers registres del conjunt de dades
display(df_partides_escacs_columnes_acp.head(10))
```

```python
# Volem fer un ACP amb només 2 components
acp=PCA(n_components=2)
# Creació del model de ACP
df_partides_escacs_model_acp=acp.fit_transform(df_partides_escacs_columnes_acp)

# Creació del dataframe amb els diferents valors obtinguts en el model ACP per a cada registre del nou conjunt de dades
df_partides_escacs_acp=pd.DataFrame(data=df_partides_escacs_model_acp,columns=['CP01','CP02'])
display(df_partides_escacs_acp.head(10))
```

```python
# Percentatge d'aportació a les components principals per a cada variable estandaritzada
df_partides_escacs_acp_total=pd.DataFrame(data=acp.components_,columns=df_partides_escacs_columnes_acp.columns,index=["CP01","CP02"])
display(df_partides_escacs_acp_total)
```

```python
# Afegir columna 'winner' al dataframe amb els diferents valors obtinguts en el model ACP per a cada registre del nou conjunt de dades
df_partides_escacs_acp=pd.concat([df_partides_escacs['winner'],df_partides_escacs_acp],axis=1)
display(df_partides_escacs_acp)
```

```python
# Percentatge de variança explicada per cadascuna de les components i del total d'elles
variansa_pct=acp.explained_variance_ratio_.round(4)
print('La primera component del ACP explica un {}% de la variança, i la segunda un {}%. Per tant, en total el model explica el {}% de la '.format(round(variansa_pct[0]*100,2),round(variansa_pct[1]*100,2),round(sum(variansa_pct)*100,2)))
```

```python
# Creació de la figura d'Anàlisi de Components Principals
fig=plt.figure(figsize=(8,8))
ax=fig.add_subplot(1,1,1) 
ax.set_xlabel('Component Principal 01',fontsize=15)
ax.set_ylabel('Component Principal 02',fontsize=15)
ax.set_title('ACP de 2 components',fontsize=20)
categories=['black','draw','white']
categories_colors=['r','g','b']
for categoria,color in zip(categories,categories_colors):
    indicesToKeep=df_partides_escacs_acp['winner']==categoria
    ax.scatter(df_partides_escacs_acp.loc[indicesToKeep,'CP01']
               ,df_partides_escacs_acp.loc[indicesToKeep,'CP02']
               ,c=color
               ,s=1)
ax.legend(categories)
ax.grid()
```
