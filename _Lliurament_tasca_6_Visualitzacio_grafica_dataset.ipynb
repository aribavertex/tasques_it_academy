# Exercici 1

### Resumeix gràficament el data set DelayedFlights.csv

-- Importació prèvia de les dades

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

vols_df=pd.read_csv("H:\_vertex\_formacio\_python - cibernarium\_03\_03_15\DelayedFlights.csv",usecols=["index","Year","Month","DayofMonth","DayOfWeek","DepTime","CRSDepTime","ArrTime","CRSArrTime","UniqueCarrier","FlightNum","ActualElapsedTime","CRSElapsedTime","AirTime","ArrDelay","DepDelay","Origin","Dest","Distance","Diverted","CarrierDelay","WeatherDelay","NASDelay","SecurityDelay","LateAircraftDelay"])
```

```python
print('Dataset de {} dades en {} variables/columnes.'.format(vols_df.shape[0],vols_df.shape[1]))
```

```python
vols_df[:10]
```

```python
vols_df.set_index('index')
```

```python
vols_df[:10]
```

-- Visualització amb la variable categòrica UniqueCarrier

```python
companyies=vols_df['UniqueCarrier'].value_counts(normalize=True)
plt.figure(figsize=(14,12))
plt.pie(companyies, labels = companyies.index,autopct='%1.1f%%',shadow = True,textprops={'fontsize': 14})
plt.title('Percentatge de vols sobre el total i per companyia',fontsize=28)
plt.show() 
```

-- Visualització amb la variable variable numèrica ArrDelay

```python
vols_df= vols_df.query('ArrDelay < 120')
x=vols_df['ArrDelay'].dropna().tolist()
plt.boxplot(x)
plt.ylabel('Retard arribada (minuts)')
plt.xlabel('ArrDelay [només vols que es retarden menys de 2 hores]')
plt.title("Boxplot de la variable ArrDelay");
plt.show()
```

-- Visualització amb una variable numèrica i una categòrica (AcumulatedDelay i UniqueCarrier)

```python
print('Les 5 companyies que acumulen de mitja els majors retards son (en hores):')
vols_df['UniqueCarrier'].describe()
vols_df.loc[:,'AcumulatedDelay']=(vols_df['ArrDelay']+vols_df['DepDelay']+vols_df['CarrierDelay']+vols_df['WeatherDelay']+vols_df['NASDelay']+vols_df['SecurityDelay']+vols_df['LateAircraftDelay'])/60
vols_df_agrupat_x_companyia_aerea_retards_acumulats=vols_df.groupby(['UniqueCarrier']).agg({'AcumulatedDelay': np.median})
vols_df_agrupat_x_companyia_aerea_retards_acumulats.sort_values(by='AcumulatedDelay', ascending=False)[0:5].plot.bar(rot=0)
```

-- Visualització amb dues variables numèriques (ArrDelay i DepDelay)

```python
x=vols_df['ArrDelay']
y=vols_df['DepDelay']
plt.scatter(x,y)
plt.ylabel('ArrDelay (minuts)')
plt.xlabel('ArrDelay (minuts)')
plt.title("DepDelay vs ArrDelay");
plt.show()
```

-- Visualització amb tres variables (ArrDelay, DepDelay i UniqueCarrier)

```python
companyies = vols_df.groupby('UniqueCarrier')

plt.figure(figsize=(22,22))
figura_numero=0

for nom, group in companyies:
    figura_numero=figura_numero+1
    plt.subplot(4,5,figura_numero)
    plt.xlim(-75, 125)
    plt.ylim(0, 250)
    x=vols_df[vols_df.UniqueCarrier==nom]['ArrDelay']
    y=vols_df[vols_df.UniqueCarrier==nom]['DepDelay']
    plt.scatter(x,y)
    plt.ylabel('DepDelay (minuts)')
    plt.xlabel('ArrDelay (minuts)')
    plt.title("Companyia " + nom)

plt.show()
```

-- Visualització amb més de tres variables (ArrDelay, DepDelay, AirTime i UniqueCarrier)

```python
#from mpl_toolkits import mplot3d

companyies = vols_df.groupby('UniqueCarrier')
fig = plt.figure(figsize=(30,30))
figura_numero=0

for nom, group in companyies:
    figura_numero=figura_numero+1
    ax=fig.add_subplot(4, 5, figura_numero, projection='3d')
    ax.set_xlim(-75, 125)
    ax.set_ylim(0, 250)    
    ax.set_zlim(0,600)
    x=vols_df[vols_df.UniqueCarrier==nom]['ArrDelay']
    y=vols_df[vols_df.UniqueCarrier==nom]['DepDelay']
    z=vols_df[vols_df.UniqueCarrier==nom]['AirTime']
    ax.scatter(x,y,z)
    ax.set_title(nom,fontweight ='bold')
    ax.set_xlabel('ArrDelay (minuts)',fontweight ='bold')
    ax.set_ylabel('DepDelay (minuts)',fontweight ='bold')
    ax.set_zlabel('AirTime (minuts)',fontweight ='bold')
    ax.grid(b=True,color='grey',linestyle ='-.',linewidth=0.3,alpha = 0.2)

    plt.show()
```
