```python
import pandas as pd
pd.options.display.max_columns = None
import random as rd
import math as m
#pd.set_option('display.max_colwidth', None)
```

```python
# Carrreguen i convertim a dataframe la taula continguda dins el fitxer/conjunt de dades anomenat games.csv
df_partides_escacs=pd.read_csv('H:\_vertex\_formacio\_python - cibernarium\_06\_06_08\games.csv',encoding="ISO-8859-15")
```

```python
# Visualització dels 5 primers registres del conjunt de dades
display(df_partides_escacs.head(5))
```

```python
# Estructura del conjunt de dades: nom, valors no nuls i tipus de columna
df_partides_escacs.info()
```

```python
# Estadístics del conjunt de dades
df_partides_escacs.describe(include="all")
```

```python
# Creació d'una mostra aleatòria simple, 'mas', de mida, n=100

n_mas=100
df_partides_escacs_mas=df_partides_escacs.sample(n_mas,replace=False)

# Comprovem si realment el tamany del nou dataframe obtingut a través de d''un mas es realment de mida correcta

if df_partides_escacs_mas.shape[0]==n_mas:
      print("El mas s'ha generat de forma correcta, te una mida de {} files/casos".format(df_partides_escacs_mas.shape[0]))
else:
      print('El mas no s''ha generat de forma correcta')
        
# Visualització de 5 elements d'aquesta mostra de 100 casos
display(df_partides_escacs_mas.head(5))
```

```python
# Creació d'una mostra sistematica, 'ms', amb un interval de 50 en 50
interval=50

# Aleatorietzar l'ordre de les files del dataframe original per a generar-ne un de nou, 'df_partides_escacs_ordre_aleatori'
df_partides_escacs_ordre_aleatori=df_partides_escacs.sample(frac=1)

# Elecció a l'atzar de d'un número que representarà quina serà la primera fila a escollir 
primera_fila_aleatoria=rd.randint(0,interval-1)
print(primera_fila_aleatoria)

# Creació ara si de la mostra sistematica de 50 en 50
df_partides_escacs_ms=df_partides_escacs_ordre_aleatori.iloc[primera_fila_aleatoria::interval]

# Comprovem si realment el tamany del nou dataframe obtingut a través de d''un mas es realment de mida correcta

if df_partides_escacs_ms.shape[0]==m.ceil(df_partides_escacs.shape[0]/interval) or df_partides_escacs_ms.shape[0]==m.floor(df_partides_escacs.shape[0]/interval):
      print("El ms s'ha generat de forma correcta, te una mida de {} files/casos".format(df_partides_escacs_ms.shape[0]))
else:
      print("El ms no s'ha generat de forma correcta")

# Visualització de 5 elements d'aquesta mostra sistemàtica
display(df_partides_escacs_ms.head(5))
```

```python
# Creació d'una mostra estratificada, 'me', de mida 150
n_me=150

# Estrats en funció de la columna de tipus de victoria 'victory_status'
victory_status_estrats=df_partides_escacs.victory_status.unique()
print(victory_status_estrats)

# Quantitat de files per estrat de 'victory_status'
victory_status_estrats_count=df_partides_escacs.groupby('victory_status').victory_status.count()
print(victory_status_estrats_count)

# Percentatge de files per estrat de 'victory_status' respecte del total de la taula original
victory_status_estrats_pctg=victory_status_estrats_count/df_partides_escacs.shape[0]
print(victory_status_estrats_pctg)

# Tamany de mostra per estrat de 'victory_status', està en float,
# pero dins el bucle el convertirem a integer per tal funcioni amb la funció sample()
victory_status_estrats_n_mostra=victory_status_estrats_pctg*n_me
print(victory_status_estrats_n_mostra)

# Creem un dataframe buit que omplirem amb la mostra estratificada
df_partides_escacs_me=pd.DataFrame()

# Inicialitzem contador per tal de asignar a cada estrat el tamany de mostra que li correspon
i=0

# Bucle per tal de generar mostra aleatoria de files dins de cada estrat
for victory_status in victory_status_estrats:
    # Selecció de files d'un determinat estrat
    df_partides_escacs_estrat=df_partides_escacs[df_partides_escacs.victory_status==victory_status]
    # Mostreig aleatori simple 'mas' en cada estrat i amb el taman de mostra que li pertoca en funció del percentatge
    df_partides_escacs_me=pd.concat([df_partides_escacs_me,df_partides_escacs_estrat.sample(int(victory_status_estrats_n_mostra[i]))])
    i=i+1

# Visualització de 5 elements d'aquesta mostra estratificada
display(df_partides_escacs_me.sample(5))
```

```python
# Creació d'una mostra balanceajada de la columna 'winner' mitjançant SMOTE
# Fer conteig de files, a pesar de que realment está balancejada la mostra,
# usem aquesta columna com a exemple d'us de l'SMOTE i usarem 'black' com que s'està submostrejant
df_partides_escacs_winner_white=df_partides_escacs[df_partides_escacs.winner=="white"]
df_partides_escacs_winner_black=df_partides_escacs[df_partides_escacs.winner=="black"]

# Tamanys de mostra en funció dels valors 'black' o 'white'
print(len(df_partides_escacs_winner_black), len(df_partides_escacs_winner_white))
```

```python
# Fem un mas en la mostra de blanques 'white' amb un tamany de mostra equivalent al de negres 'black' (en suposada minoria) 
df_winner_white_submostreig=df_partides_escacs_winner_white.sample(len(df_partides_escacs_winner_black))

# Creem el dataframe amb les dos mostres ara balancejades
df_winner_black_submostreig=pd.concat([df_partides_escacs_winner_black, df_winner_white_submostreig])

# Per tal de demostrar-ho imprimim el tamnay de mostra de totes dos submostres
print(len(df_winner_black_submostreig[df_winner_black_submostreig.winner=="white"]),df_winner_black_submostreig[df_winner_black_submostreig.winner=="black"].shape[0])

# Visualització de 5 elements d'aquesta mostra balancejada mitjançant SMOTE
df_winner_black_submostreig.sample(5)
```

```python
# Creació d'una mostra pel métode del reservoir sampling, 'rs'
n_res=250
size=200

# Llista amb les mostres/files a reservar
mostres_a_reservar=[]

for mostra in range(df_partides_escacs.shape[0]):
    # Anem afegim mostres/files a la nostra mostra de reserva 'mostres_a_reservar'
    if len(mostres_a_reservar)<n_res:
        mostres_a_reservar.append(mostra)
    # A la mostra que tinguem ja feta de reserva li substituirem una mostra per un altra amb probabilidad major
    else:
        probabilitat=n_res/(mostra+1)
        if rd.random()<probabilitat:
             mostres_a_reservar[rd.choice(range(0,n_res))]=mostra

# Filtrar les mostres de la taula original i crear-ne un dataframe nou 'df_partides_escacs_rs'
df_partides_escacs_rs=df_partides_escacs.iloc[mostres_a_reservar]

# Visualització de 5 elements d'aquesta mostra feta per eservoir sampling
display(df_partides_escacs_rs.sample(5))
```
