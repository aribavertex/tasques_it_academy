```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
pd.set_option('display.max_columns', None)
```

```python
vols = pd.read_csv('H:\_vertex\_formacio\_python - cibernarium\_04\_04_08\DelayedFlights_visualitzacio_net.csv')

vols.head()
```

```python
# Eliminar columna anomenada Unnamed: 0
vols = vols.rename(columns={"Unnamed: 0": "id"})

# Fixar com a index la columna id
vols = vols.set_index('id')

# Elimina la columna index
vols.drop('index', axis='columns', inplace=True)

vols.head()
```

```python
# Estadístics de resum de cada columna
vols.describe().round(2)
```

```python
# Eliminar algunes variables que no creiem siguin d'interés
vols = vols.drop(['ActualElapsedTime', 'CRSElapsedTime', 'Diverted', 'CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay'], axis=1)

vols.head()
```

```python
# Creació de la mostra d'entrenament i validació
mostra_entrenament, mostra_validacio = train_test_split(vols, test_size=0.2, random_state=2)

print("Tamany de mostra d'entrenament: {}".format(mostra_entrenament.shape[0]))
print("Tamany de mostra de validació: {}".format(mostra_validacio.shape[0]))

display(mostra_entrenament.head())
display(mostra_validacio.head())
```

```python
# Comparativa de diferents estadístics de columnes numériques per a les mostres d'entrenament i de validació
# display(mostra_entrenament[['AirTime','ArrDelay','DepDelay','Distance']].describe().round(2))
display(mostra_entrenament.describe(include=[np.number]).round(2))
display(mostra_validacio.describe(include=[np.number]).round(2))
```

#### La descripció retorna valors en els estadístics molt similars,sobretot a nivell de mitjes, desviacions típiques i distribució interquartílica

```python
# Comparativa de diferents estadístics de columnes categoriques per a les mostres d'entrenament i de validació
display(mostra_entrenament[['Year', 'Month', 'DayofMonth', 'DayOfWeek', 'UniqueCarrier', 'FlightNum', 'Origin', 'Dest']].describe())
display(mostra_validacio.describe(exclude=[np.number]))
```

#### La descripció retorna valors a nivell de máxims molt similars.

```python
# Fem l'estandardització de 2 variables de tipus numéric
columnes_a_estandaritzar = ['DepTime', 'ArrTime', 'AirTime', 'ArrDelay', 'DepDelay', 'Distance', 'AcumulatedDelay']
vols_col_estandaritzar = preprocessing.StandardScaler().fit(mostra_entrenament[columnes_a_estandaritzar])
vols_col_estandaritzades = vols_col_estandaritzar.transform(mostra_entrenament[columnes_a_estandaritzar])

display(vols_col_estandaritzades)
```

```python
# Afegir les columnes transformades
k = 0
columnes_a_estandaritzar_i_estandaritzades = []
for columna in columnes_a_estandaritzar:
    mostra_entrenament[columna + '_escalat'] = vols_col_estandaritzades[:,k]
    columnes_a_estandaritzar_i_estandaritzades = columnes_a_estandaritzar_i_estandaritzades + [columnes_a_estandaritzar[k]]
    columnes_a_estandaritzar_i_estandaritzades = columnes_a_estandaritzar_i_estandaritzades + [columnes_a_estandaritzar[k] + '_escalat']
    k = k + 1

print (columnes_a_estandaritzar_i_estandaritzades)
```

```python
# Visualització de la taula amb columnes estandaritzades i no estandaritzades per tal de veure i comparar
display(mostra_entrenament[columnes_a_estandaritzar_i_estandaritzades].head())
display(mostra_entrenament.head())
```

```python
# Creem una variable relacionada amb el retard acumulat per tal de categortizar la gravetat del retard

mostra_entrenament.loc[mostra_entrenament['ArrDelay'] <= 15, 'Gravetat_retard'] = 'Lleu'
mostra_entrenament.loc[(mostra_entrenament['ArrDelay'] > 15) & (mostra_entrenament['ArrDelay'] <= 60), 'Gravetat_retard'] = 'Greu'
mostra_entrenament.loc[mostra_entrenament['ArrDelay'] > 60, 'Gravetat_retard'] = 'Molt greu'

display(mostra_entrenament.head())
```

```python
# Fer el dummy de la columna 'Gravetat_retard'
gravetat_retard_dummy = pd.get_dummies(mostra_entrenament['Gravetat_retard'], prefix = 'Gravetat_retard')

gravetat_retard_dummy.head()
```

```python
# Afegir els dummies a la taula de mostra d'entrenament
mostra_entrenament = mostra_entrenament.merge(gravetat_retard_dummy, left_index=True, right_index=True)

mostra_entrenament.head()
```

```python
# Eliminació de valors nuls per tal de que sigui factible crear polinomis de regressió
tamany_mostra_abans_eliminar_nulls = mostra_entrenament.shape[0]
mostra_entrenament = mostra_entrenament.dropna()
tamany_mostra_abans_després_nulls = mostra_entrenament.shape[0]

print("S'han eliminat el {}% dels casos en la mostra d'entrenament".format(round((tamany_mostra_abans_després_nulls/tamany_mostra_abans_eliminar_nulls)*100,2)))

mostra_entrenament.head()
```

```python
# Creem un polinomi per tal d'analitzar més endavant la relació entre el retard d'arribada i la distancia recorreguda
polinomi = PolynomialFeatures(2)
polinomi_retard_distancia = polinomi.fit_transform(mostra_entrenament[['Distance_escalat','AirTime_escalat']])
polinomi_retard_distancia
```
