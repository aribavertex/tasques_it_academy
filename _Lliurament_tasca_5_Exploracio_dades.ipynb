-- Crea columnes noves:

    a. velocitat mitjana del vol,

```python
print('Càlcul de la velocitat mitja en vol en km/h. Cal tenir en compte que la distancia està inicialment en km i el temps en minuts.')
vols_df.loc[:,'AverageSpeedOnAir']=round(vols_df['Distance']/(vols_df['AirTime']/60),0)
vols_df[['index','Distance','AirTime','AverageSpeedOnAir']]
```

     b. si ha arribat tard o no...

```python
vols_df.loc[:,'ArrDelayTrueOrFalse']=vols_df['ArrDelay']
vols_df['ArrDelayTrueOrFalse']=vols_df['ArrDelayTrueOrFalse'].apply(lambda x: 'Retard' if x>0 or x<0 else 'Puntual')
vols_df[['index','ArrDelay','ArrDelayTrueOrFalse']]
```

    c. taula de les aerolínies amb més endarreriments acumulats

```python
print('Les 3 companyies que acumulen de mitja els majors retards son (en minuts):')
vols_df['UniqueCarrier'].describe()
vols_df.loc[:,'AcumulatedDelay']=vols_df['ArrDelay']+vols_df['DepDelay']+vols_df['CarrierDelay']+vols_df['WeatherDelay']+vols_df['NASDelay']+vols_df['SecurityDelay']+vols_df['LateAircraftDelay']
vols_df_agrupat_x_companyia_aerea_retards_acumulats=vols_df.groupby(['UniqueCarrier']).agg({'AcumulatedDelay': np.median})
vols_df_agrupat_x_companyia_aerea_retards_acumulats.sort_values(by='AcumulatedDelay', ascending=False)[0:3]
```

```python
vols_df_agrupat_x_companyia_aerea_retards_acumulats.sort_values(by='AcumulatedDelay', ascending=False).plot(kind='bar')
```

    c. quins són els vols més llargs?

```python
print('Els 3 vols que recorren, de mitja, les distancies més llargues son (en km):')
vols_df['FlightNum'].describe()
vols_df_agrupat_x_numero_de_vol_distancia=vols_df.groupby(['FlightNum']).agg({'Distance': np.median})
vols_df_agrupat_x_numero_de_vol_distancia.sort_values(by='Distance', ascending=False)[0:3]
```

    c. quins són els vols més endarrerits?

```python
print('Els 3 vols que, de mitja, més sendarrereixen son (en minuts):')
vols_df['FlightNum'].describe()
vols_df_agrupat_x_numero_de_vol_endarrerits=vols_df.groupby(['FlightNum']).agg({'DepDelay': np.median})
vols_df_agrupat_x_numero_de_vol_endarrerits.sort_values(by='DepDelay', ascending=False)[0:3]
```
