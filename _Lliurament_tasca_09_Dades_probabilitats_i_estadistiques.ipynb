```python
import pandas as pd
pd.options.display.max_columns=None
import statistics as stat
import numpy as np
import seaborn as sn
import scipy.stats
```

```python
# Carrreguen i convertim a dataframe la taula continguda dins el fitxer/conjunt de dades anomenat games.csv
df_partides_escacs=pd.read_csv('H:\_vertex\_formacio\_python - cibernarium\_07\_07_11\games.csv',encoding="ISO-8859-15")
```

```python
# Visualització dels 5 primers registres del conjunt de dades
display(df_partides_escacs.head(5))
```

```python
# Estructura del conjunt de dades: nom, valors no nuls i tipus de columna
df_partides_escacs.info()
```

```python
# Estadístics del conjunt de dades
df_partides_escacs.describe(include="all")
```

```python
# He escollit l'atribut 'turns', és a dir, nombre de movimens
moviments=list(df_partides_escacs.turns)

# Cálcul dels diferents estadístics de 'turns'
moviments_mitjana=stat.mean(moviments)
moviments_desv=stat.stdev(moviments)
moviments_mediana=stat.median(moviments)
moviments_moda=stat.mode(moviments)

print("Els valors dels diferents estadístics calculats son:\n\nMitjana ---------------> {}\nDesviación estandard --> {}\nMediana ---------------> {}\nModa ------------------> {}".format(round(moviments_mitjana,2),round(moviments_desv,2),round(moviments_mediana,2),moviments_moda))
```
